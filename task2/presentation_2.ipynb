```python
from typing import List
import seaborn as sns
import matplotlib.pyplot as plt


def plot_sand(iteration: int, data: List[int], sand_data: List[int], left_pnt: int, right_pnt: int, sand: int, max_left_height: int, max_right_height: int):
    sns.set_style("whitegrid")
    plt.figure(figsize=(8, 6))
    ax = sns.barplot(x=range(len(data)), y=data, color='dimgray', saturation=1, width=1)
    sns.barplot(x=range(len(sand_data)), y=sand_data, color='goldenrod', saturation=1, width=1, bottom=data) 
    plt.title(f'[Question 2] Iteration {iteration}')
    plt.xlabel('Index')
    plt.ylabel('Height')
    plt.tight_layout()

    ax.text(9.5, 4.5, f'sand={sand}', style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})
    ax.text(9.5, 4.0, f'max_left_height={max_left_height}', style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})
    ax.text(9.5, 3.5, f'max_right_height={max_right_height}', style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})

    plt.text(left_pnt, data[left_pnt] + sand_data[left_pnt] + 0.2, 'left_pnt', ha='center')
    plt.text(right_pnt, data[right_pnt] + sand_data[right_pnt] + 0.2, 'right_pnt', ha='center')

    plt.arrow(left_pnt, data[left_pnt] + sand_data[left_pnt] + 0.1, 0, -0.5, head_width=0.3, head_length=0.1, fc='black', ec='black', lw=1, width=0.05)
    plt.arrow(right_pnt, data[right_pnt] + sand_data[right_pnt] + 0.1, 0, -0.5, head_width=0.3, head_length=0.1, fc='black', ec='black', lw=1, width=0.05)

    plt.show()
```

```python
import dataclasses
from solution_2 import solve

input = [4,2,1,0,5,4,1,0,2,1,0,1]
output, states = solve(input)

stacked = [0 for _ in range(len(states[0].input))]
for iter, state in enumerate(states, 1):
    input, left_pnt, max_left_height, right_pnt, max_right_height, sand = dataclasses.astuple(state)
    if left_pnt != 0:
        stacked[left_pnt] = max_left_height - input[left_pnt]
    if right_pnt != len(input):
        stacked[right_pnt] = max_right_height - input[right_pnt]
    plot_sand(iter, input, stacked, left_pnt, right_pnt, sand, max_left_height, max_right_height)
```
